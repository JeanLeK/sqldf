```python
# Import libraries
import sqldf
import pandas as pd
import numpy as np
```

## SELECT

```python
# Load the data
url = ('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/data/tips.csv')
tips = pd.read_csv(url)
tips.head()
```

### Selecting columns from a pd.Dataframe

```python
query = """
SELECT total_bill, tip, smoker, time
FROM tips
LIMIT 5;
"""

sqldf.run(query)
```

### Adding a calculated column

```python
query = """
SELECT *, tip/total_bill as tip_rate
FROM tips
LIMIT 5;
"""

sqldf.run(query)
```

## WHERE

### Filtering a pd.Dataframe on a condition

```python
query = """
SELECT *
FROM tips
WHERE time = 'Dinner'
LIMIT 5;
"""

sqldf.run(query)
```

### Filtering with multiple conditions

#### AND conditions

```python
query = """
-- tips of more than $5.00 at Dinner meals
SELECT *
FROM tips
WHERE time = 'Dinner' AND tip > 5.00;
"""

sqldf.run(query)
```

#### OR conditions

```python
query = """
-- tips by parties of at least 5 diners OR bill total was more than $45
SELECT *
FROM tips
WHERE size >= 5 OR total_bill > 45;
"""

sqldf.run(query)
```

### Filtering null and non-null values

```python
# Create dummy pd.DataFrame
frame = pd.DataFrame({'col1': ['A', 'B', np.NaN, 'C', 'D'], 'col2': ['F', np.NaN, 'G', 'H', 'I']})
```

```python
query = """
SELECT *
FROM frame
WHERE col2 IS NULL;
"""

sqldf.run(query)
```

```python
query = """
SELECT *
FROM frame
WHERE col1 IS NOT NULL;
"""

sqldf.run(query)
```

## GROUP BY

### Counting the frequencies of aggregates

```python
query = """
SELECT sex, count(*)
FROM tips
GROUP BY sex;
"""

sqldf.run(query)
```

### Computing aggregation functions

```python
query = """
SELECT day, AVG(tip), COUNT(*)
FROM tips
GROUP BY day;
"""

sqldf.run(query)
```

### Grouping on several columns

```python
query = """
SELECT smoker, day, COUNT(*), AVG(tip)
FROM tips
GROUP BY smoker, day;
"""

sqldf.run(query)
```

## JOIN

```python
# Create dummies pd.DataFrame
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value': np.random.randn(4)})
df2 = pd.DataFrame({'key': ['B', 'D', 'D', 'E'], 'value': np.random.randn(4)})
```

### Inner join of 2 pd.DataFrame

```python
query = """
SELECT *
FROM df1
INNER JOIN df2
  ON df1.key = df2.key;
"""

sqldf.run(query)
```

### Left join of 2 pd.DataFrame

```python
query = """
-- show all records from df1
SELECT *
FROM df1
LEFT OUTER JOIN df2
  ON df1.key = df2.key;
"""

sqldf.run(query)
```

### Right join and Full join are not currently supported by SQLite3

## UNION

```python
# Create dummies pd.DataFrame
df1 = pd.DataFrame({'city': ['Chicago', 'San Francisco', 'New York City'], 'rank': range(1, 4)})
df2 = pd.DataFrame({'city': ['Chicago', 'Boston', 'Los Angeles'], 'rank': [1, 4, 5]})
```

### Union of 2 pd.DataFrames with duplicates

```python
query = """
SELECT city, rank
FROM df1
UNION ALL
SELECT city, rank
FROM df2;"""

sqldf.run(query)
```

### Union of 2 pd.DataFrames without duplicates

```python
query = """
SELECT city, rank
FROM df1
UNION
SELECT city, rank
FROM df2;"""

sqldf.run(query)
```

## TOP N rows with offset

```python
query = """
SELECT * FROM tips
ORDER BY tip DESC
LIMIT 10 OFFSET 5;
"""

sqldf.run(query)
```

## UPDATE

### Update a column with a calculated column

```python
query = """
UPDATE tips
SET tip = tip*2
WHERE tip < 2;
"""

sqldf.run(query)
tips.head()
```

## DELETE

### Remove some rows from a pd.DataFrame based on a condition

```python
print("Initial number of rows:", tips.shape[0])

query = """
DELETE FROM tips
WHERE tip > 9;
"""

sqldf.run(query)
print("Final number of rows:", tips.shape[0])
```
